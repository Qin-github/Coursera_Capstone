```python
import pandas as pd
import numpy as np
```

## read data 

```python
df = pd.read_csv('table.csv') 
```

```python
df
```

## create the dataframe called df

### Step 1 rename the column "Postcode" to "PostalCode"

```python
df.rename(columns = {"Postcode":"PostalCode"}, inplace = True)
```

```python
df
```

### Step 2 Remove the "Not assigned" cells from "Borough" column

```python
df.replace({"Borough":"Not assigned"}, np.nan, inplace = True)
df = df.dropna()
df
```

### Step 3 Replace the "Not assigned" in "Neighbourhood" column 

```python
df.Neighbourhood[df["Neighbourhood"] == "Not assigned"] = df["Borough"]
df.head(10)
```

```python
df['Neighbourhood'] = df[['PostalCode', 'Borough', 'Neighbourhood']].groupby(['PostalCode', 'Borough'])['Neighbourhood'].transform(lambda x: ','.join(x))
df = df[['PostalCode', 'Borough', 'Neighbourhood']].drop_duplicates()
df
```

### The final dataframe

```python
df = df[['PostalCode', 'Borough', 'Neighbourhood']]
df
```
